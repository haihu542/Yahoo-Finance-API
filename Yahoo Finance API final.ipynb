```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
pd.options.display.max_rows = None
```

```python
def get_data(ticker_symbol):
    print('Scroll down for financials and stock actions')
    pull_ticker = yf.Ticker(ticker_symbol)
    historical_data = pull_ticker.history(period="2y")  # data for the last year   
    fig = go.Figure()
    column_names = historical_data.columns
    fig.add_trace(go.Scatter(x=historical_data.index, y=historical_data["Volume"], mode='lines', name='LUNR'))
    fig.show()
    financials = pull_ticker.financials
    print("\nFinancials:")
    print(financials)
    print('/n')
    print('/n')
    # Fetch stock actions like dividends and splits
    actions = pull_ticker.actions
    print("\nStock Actions:")
    print(actions)
    
```

```python
get_data("AAPL")
```
